## Distribution of `VALUE`, `% ACTIVATION` and `% INHIBITION` among assays

Note: In Assay 8 there's `Value` instead of `VALUE` column. It will be convenient co capitalize all column names to keep the consistency

```python
def get_distributions(d, colname, color):
    to_plot = list()
    for key, assay in d.items():
        if colname in assay:
            to_plot.append(key)

    fig, ax = plt.subplots(1, len(to_plot), figsize =(25, 5))
    for assay, subplot in zip(to_plot, ax):
        if colname in d[assay]:
            sns.histplot(d[assay][colname], x=d[assay][colname], kde=True, ax=subplot, color=color).set(title=assay)
```

```python
print(f'Distribution of VALUE among assays')
get_distributions({k: d[k] for k in list(d)[:4]}, 'VALUE', 'purple')
get_distributions({k: d[k] for k in list(d)[4:]}, 'VALUE', 'purple')
```

```python
print(f'Distribution of % ACTIVATION among assays')
get_distributions(d, '% ACTIVATION', 'green')
```

```python
print(f'Distribution of % INHIBITION among assays')
get_distributions(d, '% INHIBITION', 'blue')
```

## Correlation between `VALUE`, `% ACTIVATION` and `% INHIBITION` among assays

Note: In Assays 1-3 the same compounds appear twice, which is not the case in Assays 4-8 (Compound ID is unique). Therefore, compounds in 4-8 will be duplicated while joining.

```python
# adding suffixes
for i in range(1,9):
    d[f'Assay {i}'] = d[f'Assay {i}'].add_suffix(f' {i}')
```

```python
for i in range(1,4):  
    d[f'Assay {i}']['ID'] = (d[f'Assay {i}'][f'ASSAY {i} - CMPD ID {i}'].astype(str) + d[f'Assay {i}'][f'BARCODE ASSAY PLATE {i}'].str[:-2]) + d[f'Assay {i}'][f'BARCODE ASSAY PLATE {i}'].str[-1]
```

```python
# merging Assays 1-3 wrt ID created
combined_df = pd.merge(d['Assay 1'], d['Assay 2'], on='ID')
combined_df = pd.merge(combined_df , d['Assay 3'], on='ID')
combined_df.head()
```

```python
# merging Assays 4-9 with combined_df wrt compound ID 
for i in range(4, 9):
    combined_df = pd.merge(combined_df , d[f'Assay {i}'], left_on='ASSAY 1 - CMPD ID 1', right_on=f'ASSAY {i} - CMPD ID {i}')
```

```python
combined_df = combined_df.rename(columns={'ASSAY 1 - CMPD ID 1':'CMPD ID'})
combined_df.columns
```

```python
len(combined_df.index)
```

```python
colnames = ['CMPD ID']
colnames.extend([f'% ACTIVATION {i}' for i in range(1,4)])
colnames.extend([f'% INHIBITION {i}' for i in range(4,9)])
colnames.extend(['% ACTIVATION 8'])
colnames_values = ['CMPD ID']
colnames_values.extend([f'VALUE {i}' for i in range(1,9)])

res = combined_df[colnames]
display(res.head())
res_values = combined_df[colnames_values]
display(res_values.head())
```
